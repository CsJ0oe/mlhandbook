# Introduction To Data Analysis

```python
import numpy as np
import pandas as pd
```

## Loading data

```python
df_train = pd.read_csv('../../datasets/ames_housing/train.csv')
```

```python
# Print dataset shape (rows and columns)
print(f"Dataset shape: {df_train.shape}")
```

```python
# Print a concise summary of the dataset
df_train.info()
```

```python
# Show the 10 first samples
df_train.head(n=10)
```

```python
# Show 10 random samples
df_train.sample(n=10)
```

```python
# Select numerical columns
df_numerical = df_train.select_dtypes(include=[np.number])

print(f"{len(df_numerical.columns)} numerical columns")
print(df_numerical.columns)
```

```python
# Select categorical columns
df_categorical = df_train.select_dtypes(include=[np.object])

print(f"{len(df_categorical.columns)} numerical columns")
print(df_categorical.columns)
```

```python
# Print descriptive statistics for all numerical columns
df_train.describe()
```

```python
# Print all values for a categorical column
df_train["SaleCondition"].value_counts()
```

## Missing data

```python
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum() * 100/df_train.isnull().count()).sort_values(ascending=False)

df_missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
df_missing_data.head(20)
```
